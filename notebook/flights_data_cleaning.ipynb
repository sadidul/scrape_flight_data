```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("flights_data.csv")
df.head(10)
```

```python
df.info()
```

```python
# Convert columns to datetime
df['scrape_date'] = pd.to_datetime(df['scrape_date'])
df['travel_date'] = pd.to_datetime(df['travel_date'])

# Calculate the difference in days
df['booking_before_days'] = (df['travel_date'] - df['scrape_date']).dt.days
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['stoppage'].value_counts()
```

```python
# For duration (handle missing values)
df['duration_mins'] = (
    df['duration']
    .str.extract(r'(\d+) hr (\d+) min')  # Split into hours and minutes
    .fillna(0)                           # Replace missing values with 0
    .astype(int)                         # Convert to integers
    .mul([60, 1])                        # Multiply hours by 60, minutes by 1
    .sum(axis=1)                         # Sum hours+minutes
)

# For stoppage (handle "Nonstop" and missing values)
df['stoppage'] = (
    df['stoppage']
    .str.replace('Nonstop', '0', case=False)  # Case-insensitive replacement
    .str.extract('(\d+)', expand=False)       # Extract numeric part
    .fillna(0)                                # Fill missing with 0
    .astype(int)                              # Convert to integer
)
```

```python
df.head(15)
```

```python
df.drop(columns=['duration', 'booking_before'], inplace=True)
```

```python
df.head()
```

```python
# Rename the column (case-sensitive operation)
df = df.rename(columns={'ticket_price_BDT': 'ticket_price_bdt'})

# Reorder columns to move ticket_price_bdt to last position
new_column_order = [col for col in df.columns if col != 'ticket_price_bdt'] + ['ticket_price_bdt']
df = df[new_column_order]
```
